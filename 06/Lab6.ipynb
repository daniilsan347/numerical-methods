xi           :   -2,   -1, 0,  3  
f(xi)        :   -4,    3, 2, 11  
Задані точки : -1.5, -0.5, 1,  2

```python
def custom_langrage(x_list, y_list, x):
    y = 0
    for j in range(len(y_list)):
        p1 = 1
        p2 = 1
        for i in range(len(x_list)):
            if i == j:
                continue
            p1 *= x - x_list[i]
            p2 *= x_list[j] - x_list[i]
        y += y_list[j] * p1 / p2
    return y
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_list = np.array([-2, -1, 0,  3])
y_list = np.array([-4,  3, 2, 11])

xz = np.array([-1.5, -0.5, 1., 2.])
yz = [custom_langrage(x_list, y_list, x) for x in xz]

x = np.linspace(x_list.min(), x_list.max(), 100)
y = [custom_langrage(x_list, y_list, i) for i in x]

plt.title('Custom Langrage')
plt.plot(x, y, 'r', x_list, y_list, 'bo', xz, yz, 'g^')
plt.grid(True)
```

```python
from scipy.interpolate import lagrange

L = lagrange(x_list, y_list)
plt.title('Scipy Langrage')
plt.plot(x, L(x), 'g', x_list, y_list, 'bo')
plt.grid(True)
```
