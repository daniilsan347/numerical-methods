```python
# Таблиця 1
tab1_x  = 2.4 
tab1_dx = 0.2
tab1_i  = 11
tab1_yi = [3.526, 3.782, 3.945, 4.043, 4.104, 4.155,
           4.222, 4.331, 4.507, 4.775, 5.159, 5.683]

# Таблиця 2
tab2_x  = 1.5
tab2_dx = 0.5
tab2_i  = 11
tab2_yi = [10.517, 10.193, 9.807, 9.387, 8.977, 8.637,
           8.442, 8.482, 8.862, 9.701, 11.132, 13.302]
```

```python
def find_deltas(yi):
    """Пошук різниць між значеннями функції"""
    deltas = []
    for i in range(1, len(yi)):
        deltas.append(abs(yi[i] - yi[i - 1]))
    return deltas
```

```python
def find_all_dy_for_i(dy_list, i):
    """Пошук різниць для i-го порядку"""
    deltas = []
    for col in dy_list:
        try:
            deltas.append(col[i])
        except IndexError:
            break
    deltas.pop(0)
    return deltas
```

```python
import numpy as np

# Пошук кінцевих різниць для таблиці 1
tab1_deltas = [tab1_yi]
for i in range(tab1_i):
    tab1_deltas.append(find_deltas(tab1_deltas[i]))

# Виведення заголовку таблиці 1
print(f'{"Таблиця 1 кінцевих різниць":^108}')

# Виведення деліменатора таблиці 1
print(f'+----+-----+-------+', end='')
for i in range(1, tab1_i+1):
    print(f'-------+', end='')
print()

print(f'|  i |  xi |    yi |', end='')
for i in range(1, tab1_i+1):
    print(f'  Δ{i}yi |' if i < 10 else f' Δ{i}yi |', end='')
print()

# Виведення деліменатора таблиці 1
print(f'+----+-----+-------+', end='')
for i in range(1, tab1_i+1):
    print(f'-------+', end='')
print()

# Виведення значень таблиці 1
tab1_xi = np.arange(tab1_x, tab1_x + tab1_dx * tab1_i + 0.001, tab1_dx)
for i in range(tab1_i + 1):
    print(f'| {i:2} | {tab1_xi[i]:.1f} |', end='')
    for col in tab1_deltas:
        try:
            print(f' {col[i]:.3f} |', end='')
        except IndexError:
            print('       |', end='')
    print()

# Виведення деліменатора таблиці 1
print(f'+----+-----+-------+', end='')
for i in range(1, tab1_i+1):
    print(f'-------+', end='')
print()
```

```python
# Пошук кінцевих різниць для таблиці 2
tab2_deltas = [tab2_yi]
for i in range(tab2_i):
    tab2_deltas.append(find_deltas(tab2_deltas[i]))

# Виведення заголовку таблиці 2
print(f'{"Таблиця 2 кінцевих різниць":^120}')

# Виведення деліменатора таблиці 2
print(f'+----+-----+--------+', end='')
for i in range(1, tab1_i+1):
    print(f'--------+', end='')
print()

print(f'|  i |  xi |     yi |', end='')
for i in range(1, tab2_i+1):
    print(f'   Δ{i}yi |' if i < 10 else f'  Δ{i}yi |', end='')
print()

# Виведення деліменатора таблиці 2
print(f'+----+-----+--------+', end='')
for i in range(1, tab2_i+1):
    print(f'--------+', end='')
print()

# Виведення значень таблиці 2
tab2_xi = np.arange(tab2_x, tab2_x + tab2_dx * (tab2_i + 1), tab2_dx)
for i in range(tab2_i + 1):
    print(f'| {i:2} | {tab2_xi[i]:.1f} |', end='')
    for col in tab2_deltas:
        try:
            print(f' {col[i]:6.3f} |', end='')
        except IndexError:
            print('        |', end='')
    print()

# Виведення деліменатора таблиці 2
print(f'+----+-----+--------+', end='')
for i in range(1, tab1_i+1):
    print(f'--------+', end='')
print()
```

```python
def derivative1(diy, h):
    """Пошук похідної 1-го порядку"""
    result = 0
    for i in range(len(diy)):
        result += (-1) ** i * (diy[i] / (i+1))
    return result / h
```

```python
import scipy as sp

def fact(x):
    if x == 0:
        return 1
    return x * fact(x-1)

def make_f(i):
    def f(q):
        if i == 0:
            return 1
        
        result = q
        for j in range(i-1):
            result *= (q - (j+1))
        return result / fact(i)
    return f

def derivative2(diy, h):
    """Пошук похідної 2-го порядку"""
    result = 0
    for i in range(len(diy)):
        result += diy[i] * sp.misc.derivative(make_f(i+1), 0, dx=0.001, n=2)
    return result / h**2
```
